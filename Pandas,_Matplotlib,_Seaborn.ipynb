```
%matplotlib inline

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
data = pd.read_csv('fitness.csv')
print(data.head())
```

```
print(data.describe())
```

```
gender1 = data[data['Gender'] == 1]
print(gender1.describe())
```

```
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=[9, 9])
plt.scatter(data['Height'], data['Weight'])
plt.xlabel('Height')
plt.ylabel('Weight')
plt.grid(True)
```

```
data.plot.scatter(x='Height', y='Weight', figsize=[9, 9])
plt.grid(True)
```

```
plt.figure(figsize=[9, 9])
sns.scatterplot(x='Height', y='Weight', hue='Gender', data=data)
plt.grid(True)
```

```
data.boxplot(column='Weight', by='Exercise', figsize=[9, 9])
```

```
plt.figure(figsize=[9,9])
sns.set_style("darkgrid")
sns.boxplot(data=data, x='Exercise', y='Weight')
```
