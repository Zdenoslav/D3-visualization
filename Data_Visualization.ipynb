```python
df_data = pd.read_csv('data.csv',encoding='latin1')
```

```python
df_data.columns = ['Unnamed: 0', 'Track.Name', 'Artist.Name', 'Genre', 'Beats.Per.Minute',
       'Energy', 'Danceability', 'Loudness', 'Liveness', 'Valence',
       'Length', 'Acousticness', 'Speechiness', 'Popularity'] ##changing the name of columns.
```

```python
df_data.corr() #Coorelation of columns
```

```python
print(df_data.describe())
```

```python

## writing a function which will scale all the required columns so, that they are in one range 0 to 1
min_max_scaler = preprocessing.MinMaxScaler()
for a in df_data.iloc[:,4:].columns:
    x = df_data[[a]].values.astype(float)
    df_data[a] = min_max_scaler.fit_transform(x)
```

```python
df_data
```

```python
## choosing the columns which have the maximum effect on the ranking of song.
fig = go.Figure(data=[
    go.Bar(name='Energy', x=df_data['Track.Name'], y=df_data['Energy'],marker_color='rgb(117, 132, 173)'),
    go.Bar(name='Danceability', x=df_data['Track.Name'], y=df_data['Danceability'],marker_color='rgb(49, 69, 112)'), #3
    go.Bar(name='Liveness', x=df_data['Track.Name'], y=df_data['Liveness'],marker_color='rgb(174, 209, 214)'),#2
    go.Bar(name='Loudness', x=df_data['Track.Name'], y=df_data['Loudness'],marker_color='rgb(224, 152, 115)'),
    go.Bar(name='Valence', x=df_data['Track.Name'], y=df_data['Valence'],marker_color='rgb(163, 127, 111)'),
    go.Bar(name='Speechiness', x=df_data['Track.Name'], y=df_data['Speechiness']),

],)
# Change the size and barmode activated, as data is too much so, we are increasing the width of data to adjust it.
fig.update_layout(
    autosize=True,
    height=600,
    title="TOP 50 SPOTIFY SONGS",
    margin=dict(
        l=10,
        r=10,
        b=200,
        t=100,
        pad=4
    ),
    font=dict(
        family="Amiri",
    ),
    paper_bgcolor="LightSteelBlue",
        bargap=0.4,
    barmode='stack'
)
fig.show()
```

```python
fig.write_html("visual1.html")
```
